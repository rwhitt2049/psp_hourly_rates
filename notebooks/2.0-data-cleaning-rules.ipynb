```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv(r"../data/raw/hourly_power_rates.csv", parse_dates=["date"])
```

```python
df.columns
```

```python
start_times = df["Time of Day (CT)"].str.extract(r"(\d{1,2}:\d{2} [A|P]M)", expand=False)
```

```python
type(start_times)
```

```python
date = df["date"]
```

```python
df["start_datetime"] = date.combine(start_times, lambda date, time: dt.datetime.combine(date, pd.Timestamp(time).time()))
```

```python
type(start_times)
```

```python
df
```

```python
end_times = df["Time of Day (CT)"].str.extract(r"\d{1,2}:\d{2} [A|P]M - (\d{1,2}:\d{2} [A|P]M)", expand=False).astype("datetime64[h]") - pd.Timedelta(seconds=1)
```
